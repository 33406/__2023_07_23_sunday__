```python
import random
#days = [num for num in range(1,31)]
days = list(range(1,31))
celsius = [round(random.uniform(24.3,37.5),ndigits = 1) for _ in range(1,31)]
figure = plt.fligure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(days,celsius,'r--o')
axes.legend()
axes.set_title("Weather")
axes.set_xlebel("Days")
axes.set_ylebel("Celsius")
plt.show()
```
