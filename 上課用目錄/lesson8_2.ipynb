```python
import numpy as np
import matplotlib.pyplot as plt

index = np.arange(5)
value = np.array([5,7,3,4,6]) * 1000
figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.bar(index,value)
axes.set_xticks(index,["apple","orange","pineapple","mango","coconut"])
axes.grid(axis='y')
for i in index:
    print(i,value[i],end="")
    axes.annotate(str(value[i]),str([i-0.2,value[i]]))

plt.show()
```

```python
import numpy as np
import pandas as pd

score = np.random.randint(55,high=100,size=(25,5))
students = pd.DataFrame(score,index=range(1,26),columns=["國文","英文","數學","社會","自然"])
students
```

```python
#使用subscript->["字串"]選欄
#使用subscript->[1:5]選列
students["國文"]
```

```python
#前五位學生
students.loc[1:5],students.iloc[0:5]
```

```python
one = students.loc[1] #不建議的寫法
one.sum()

students.loc[1].sum()#建議的寫法
```

```python
for i in range(1,26):
    print(students.loc[i].sum(),end=" ")
```

```python
students["總分"] = students.sum(axis=1)
students
```

```python
students.loc[1:25,["國文","英文","數學","社會","自然"]]
students["平均"] = students.iloc[:,:5].mean(axis=1) #dataframe
students
```
